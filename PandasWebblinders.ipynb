```python



# Read the uploaded file
df = pd.read_csv(io.BytesIO(uploaded['Iris.csv']))
print(df.head())  # Display first 5 rows

# Load the Iris dataset
iris_df = pd.read_csv("Iris.csv")

# 1. Display first 5 rows
print("First 5 rows:\n", iris_df.head())

# 2. Display last 5 rows
print("\nLast 5 rows:\n", iris_df.tail())

# 3. Shape of the dataset (rows, columns)
print("\nShape of dataset:", iris_df.shape)

# 4. Columns in the dataset
print("\nColumns in dataset:", iris_df.columns)

# 5. Data types of each column
print("\nData types:\n", iris_df.dtypes)

# 6. Checking for null values
print("\nMissing values:\n", iris_df.isnull().sum())

# 7. Basic statistical summary
print("\nStatistical Summary:\n", iris_df.describe())

# 8. Display dataset information
iris_df.info()

# 9. Unique values in Species column
print("\nUnique species:", iris_df["Species"].unique())

# 10. Count of each unique species
print("\nSpecies count:\n", iris_df["Species"].value_counts())

# 11. Selecting a single column
print("\nSepal Length:\n", iris_df["SepalLengthCm"].head())

# 12. Selecting multiple columns
print("\nSelected Columns:\n", iris_df[["SepalLengthCm", "PetalLengthCm"]].head())

# 13. Filtering rows where SepalLengthCm > 5.5
print("\nFiltered Rows (SepalLengthCm > 5.5):\n", iris_df[iris_df["SepalLengthCm"] > 5.5].head())

# 14. Selecting rows using index slicing
print("\nIndexed Selection:\n", iris_df.iloc[1:50:4])

# 15. Sorting by SepalWidthCm
sorted_df = iris_df.sort_values(by="SepalWidthCm")
print("\nSorted by SepalWidthCm:\n", sorted_df.head())

# 16. Resetting index
sorted_df = sorted_df.reset_index(drop=True)
print("\nReset Index after Sorting:\n", sorted_df.head())

# 17. Grouping by Species and getting mean
print("\nGrouped by Species:\n", iris_df.groupby("Species").mean())

# 18. Adding a new column (Sepal Area)
iris_df["SepalArea"] = iris_df["SepalLengthCm"] * iris_df["SepalWidthCm"]
print("\nWith New Column SepalArea:\n", iris_df.head())

# 19. Dropping a column (PetalWidthCm)
iris_df_dropped = iris_df.drop(columns=["PetalWidthCm"])
print("\nAfter Dropping 'PetalWidthCm':\n", iris_df_dropped.head())

# 20. Renaming SepalLengthCm to Sepal_Length
iris_df_renamed = iris_df.rename(columns={"SepalLengthCm": "Sepal_Length"})
print("\nAfter Renaming Column:\n", iris_df_renamed.head())


```
